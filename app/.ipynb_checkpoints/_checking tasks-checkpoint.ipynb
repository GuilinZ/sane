## PCA for bbox

```python
### applying PCA

import numpy as np
from sklearn.decomposition import PCA

#X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])

rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T

plt.scatter(X[:,0], X[:,1])
plt.axis('equal')
plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(X)  

#print(pca.explained_variance_ratio_)  
print(pca.components_)

def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1])

for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');    
```

```python
#considering all points

X_pca = pca.transform(X)

X_pc1_minmax = [X_pca[:,0].min(), X_pca[:,0].max()]
X_pc2_minmax = [X_pca[:,1].min(), X_pca[:,1].max()]

X_pca_bound = np.array([[X_pc1_minmax[0], X_pc2_minmax[0]], [X_pc1_minmax[0], X_pc2_minmax[1]], [X_pc1_minmax[1], X_pc2_minmax[0]], [X_pc1_minmax[1], X_pc2_minmax[1]]])
X_bound = pca.inverse_transform(X_pca_bound)

#plotter
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.8)
plt.axis("equal")
plt.scatter(X_pca_bound[:, 0], X_pca_bound[:, 1], s=100, c="black")

plt.xlabel("PC 1 Axis")
plt.ylabel("PC 2 Axis")
plt.show()
```

```python
#ignoring outliers
X_pca = pca.transform(X)

#outlier bounds
IQR = np.quantile(X_pca, 0.75, 0) - np.quantile(X_pca, 0.25, 0)
IQR_thres = 2 #1.5
outlier_bounds = np.array([np.quantile(X_pca, 0.5, 0) - IQR_thres*IQR, np.quantile(X_pca, 0.5, 0) + IQR_thres*IQR])

#PCA1
PCA1_filter = np.logical_and(X_pca[:,0] >= outlier_bounds[0,0], X_pca[:,0] <= outlier_bounds[1,0])
PCA2_filter = np.logical_and(X_pca[:,1] >= outlier_bounds[0,1], X_pca[:,1] <= outlier_bounds[1,1])

X_pca_filtered = X_pca[np.logical_and(PCA1_filter, PCA2_filter),:]

X_pc1_minmax = [X_pca_filtered[:,0].min(), X_pca_filtered[:,0].max()]
X_pc2_minmax = [X_pca_filtered[:,1].min(), X_pca_filtered[:,1].max()]

X_pca_bound_filtered = np.array([[X_pc1_minmax[0], X_pc2_minmax[0]], [X_pc1_minmax[0], X_pc2_minmax[1]], [X_pc1_minmax[1], X_pc2_minmax[0]], [X_pc1_minmax[1], X_pc2_minmax[1]]])
X_bound_filtered = pca.inverse_transform(X_pca_bound_filtered)


#plotter
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.8)
plt.axis("equal")
plt.scatter(X_pca_bound[:, 0], X_pca_bound[:, 1], s=100, c="grey")
plt.scatter(X_pca_bound_filtered[:, 0], X_pca_bound_filtered[:, 1], s=100, c="black")

#plotting bounds
plt.axhline(y=outlier_bounds[0, 1])
plt.axhline(y=outlier_bounds[1, 1])
plt.axvline(x=outlier_bounds[0, 0])
plt.axvline(x=outlier_bounds[1, 0])

plt.xlabel("PC 1 Axis")
plt.ylabel("PC 2 Axis")
plt.show()
```

```python
# plot data and bounds
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.axis('equal');    
plt.scatter(X_bound[:, 0], X_bound[:, 1], s=100, c="grey")
plt.scatter(X_bound_filtered[:, 0], X_bound_filtered[:, 1], s=100, c="black")
plt.show()
```
